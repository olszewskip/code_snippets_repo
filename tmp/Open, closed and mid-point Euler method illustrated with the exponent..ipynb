#### Find an approximate solution *y(t)* using different Euler-methods and time-steps *h*:
\begin{align}
& y'(t) = A \cdot  y(t) ,\\
& t \in \left[ t_0, t_1, \ldots, t_{N-1} \right]
 \;, \quad
 h\equiv t_{i+1}-t_i
 \;,\quad 
 y(t_0) \equiv y_0
\end{align}


##### *Open* method:
\begin{align}
&y(t_{i+1}) = y(t_i) + y'(t_i) \cdot h \\
&y(t_{i+1}) = (1 + h \cdot A) \,y(t_i)
\end{align}

##### *Closed* method:
\begin{align}
& y(t_{i+1}) = y(t_i) + y'(t_{i+1}) \cdot h \\
& y(t_{i+1}) = \frac{1}{1 - h \cdot A} y(t_i)
\end{align}
 
##### *Mid-point* method:
\begin{align}
&\text{e.g.}\quad y_1 = (1 + A\cdot h) \,y(t_0) \\
&y_{i+2} = y_{i} + 2h \cdot y'(t_{i+1}) \\
&y_{i+2} = y_{i} + 2h \cdot A \cdot y(t_{i+1})
\end{align}

#### Take for example
\begin{align}
&A \in \{-10,\, 10 \} \\
&  y_0 = 1 \\
&  t_0 = 0 \;,\quad t_1 = 1 \\
&  h \in \{0.05, \,0.101,\, 0.2\}
\end{align}

---

```python
from math import exp
import matplotlib.pyplot as plt
```

---
Generate time-points *t*.

```python
def time_steps(h, t_0=0, t_1=1):
    step_count = int( (t_1 - t_0) / h )
    return [h * i for i in range(0,step_count+1)]
```

Find values of the exact solution $exp(A\cdot t)$.

```python
def y_exact(time_steps, A):
     return [exp(A * t) for t in time_steps]
```

---
Define the Euler methods as functions $y_i \rightarrow y_{i+1}$ parametrized by $A$ and $h$.

```python
def open_(A, h):
    def step1(y):
        y_next = y * (1 + A * h)
        return y_next
    return step1
```

```python
def closed(A, h):
    def step1(y):
        y_next = y / (1 - A * h)
        return y_next
    return step1
```

```python
def midpoint(A, h):
    def step2(y_0, y_1):
        y_next = y_0 + 2 * h * ( A * y_1) 
        return y_next
    return step2
```

Given an Euler method, some chosen $y_0$, and knowing the number of time-steps, return series of the approximate values $y(t)$. 

```python
def y_approx(euler, time_steps, y_0=1, euler0=None):
    
    if euler.__name__ == 'step2':
        y_1 = euler0(y_0)
        y = [y_0, y_1]
        for _ in time_steps[2:]:
            y_next = euler(y[-2], y[-1])
            y.append(y_next)
    else:
        y = [y_0]
        for _ in time_steps[1:]:
            y_next = euler(y[-1])
            y.append(y_next)
    
    return y
```

---
Plot results without the *midpoint* method.

```python
for A in [-10, 10]:
    for h in [0.105, 0.2, 0.3]:
        _, _ = plt.subplots(figsize=(6,4))
        t = time_steps(h)
        plt.plot(t, y_exact(t, A), "--")
        plt.plot(t, y_approx(open_(A,h), t), "-o")
        plt.plot(t, y_approx(closed(A,h), t), ":d")
        #plt.plot(t, y_approx(midpoint(A,h), t, euler0=open_(A,h)), "-.s")
        plt.xlabel("t")
        plt.ylabel("$y(t)$")
        plt.title(f"A={A}, h={h}")
        plt.legend(['Exact', 'Open', 'Closed'])
        plt.show()
```

Plot results without the *closed* method.

```python
for A in [-10, 10]:
    for h in [0.105, 0.2, 0.3]:
        _, _ = plt.subplots(figsize=(6,4))
        t = time_steps(h)
        plt.plot(t, y_exact(t, A), "--")
        plt.plot(t, y_approx(open_(A,h), t), "-o")
        #plt.plot(t, y_approx(closed(A,h), t), ":d")
        plt.plot(t, y_approx(midpoint(A,h), t, euler0=open_(A,h)), "-.s")
        plt.xlabel("t")
        plt.ylabel("$y(t)$")
        plt.title(f"A={A}, h={h}")
        plt.legend(['Exact', 'Open', 'Midpoint'])
        plt.show()
```

Using $|A\cdot h | > 1$ gives poor or completely uninteresting results. 

Plot all methods with smaller time-step *h* and a sligtly narrower time-span.

```python
h0 = 0.01
t_0 = 0
t_1 = 0.5

for A in [-10, 10]:
    for h in [h0]:
        _, _ = plt.subplots(figsize=(8,6))
        t = time_steps(h, t_0 = t_0, t_1=t_1)
        plt.plot(t, y_exact(t, A), "--")
        plt.plot(t, y_approx(open_(A,h), t), "-o")
        plt.plot(t, y_approx(closed(A,h), t), ":d")
        plt.plot(t, y_approx(midpoint(A,h), t, euler0=open_(A,h)), "-.s")
        plt.xlabel("t")
        plt.ylabel("$y(t)$")
        plt.title(f"A={A}, h={h}")
        plt.legend(['Exact', 'Open', 'Closed','Midpoint'])
        plt.show()
```

The *midpoint* is unstable for negative A but it is the most accurate of the three methods for positive A.
